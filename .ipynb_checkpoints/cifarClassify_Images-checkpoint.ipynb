### cifarClassify_Images

Veri setinin yüklenmesi ve yüklenen veri kümelerinin veri türü.

```python
from keras.datasets import cifar10
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
#veritipleri
print(type(x_train))
print(type(y_train))
print(type(x_test))
print(type(y_test))
```

Yüklenen veri kümelerinin boyutu

```python
#boyutlar
print('x_train shape:', x_train.shape) #4D array 50,000 rows 32x32 pixel image with depth = 3 visible wave lenghts (RGB)
print('y_train shape:', y_train.shape) #2D array 50,000 rows and 1 column
print('x_test shape:', x_test.shape) #4D array 10,000 rows 32x32 pixel image with depth = 3 visible wave lenghts (RGB)
print('y_test shape:', y_test.shape) #2D array 10,000 rows and 1 column
```

Veri setindeki ilk goruntu ve label

```python
#training datadaki ilk resim
x_train[0]
```

```python
import matplotlib.pyplot as plt
img = plt.imshow(x_train[0])
```

```python
#goruntunun labeli
#0 = airplane
#1 = automobile
#2 = bird
#3 = cat
#4 = deer
#5 = dog
#6 = frog
#7 = horse
#8 = ship
#9 = truck
print('The label is:', y_train[0])
```

Etiketleri sinir ağına girmek için 10 numaradan oluşan
bir kümeye dönüştürmek için one-hot encoding
kullanıldı . Sayılar görüntüleri sınıflandırmak için etiket
sayısına karşılık gelir.

```python
#One-Hot Encoding 
from keras.utils import to_categorical
y_train_one_hot = to_categorical(y_train)
y_test_one_hot = to_categorical(y_test)
```

One-hot encoding sonrası y_train (eğitim veri seti) veri
kümesindeki tüm etiketler:

```python
print(y_train_one_hot)
```

```python
print('The one hot label is:', y_train_one_hot[0])
```

- Görüntülerdeki piksellerin 0 ile 1 arasında bir değer
olacak şekilde normalleştirilmesi.
- Normalde 0 ile 255 arasında değerlerdir. Bu sinir ağının
ilerlemesine yardımcı olur.

```python
x_train = x_train / 255
x_test = x_test / 255
```

Convolution Neural Network Model Oluşturma

Sequential() kullanarak mimariyi oluşturmak:

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D

model = Sequential() 

model.add(Conv2D(32, (5, 5), activation='relu', input_shape=(32,32,3))) 

model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (5, 5), activation='relu'))

model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
model.add(Dense(1000, activation='relu')) 
model.add(Dense(10, activation='softmax')) 
```

Model derlenir.

```python
model.compile(loss='categorical_crossentropy', 
              optimizer='adam',
              metrics=['accuracy'])
```

Train için fit() yöntemini kullanarak model eğitildi.
• Modeli, batch_size=256, epochs=10 olan eğitim verileri üzerinde eğitildi ve
verilerin %70'i eğitimde kullanılarak ve diğer %30'u doğrulama (validation)
olarak ayrıldı.
• Batch : Tek bir batch içinde mevcut toplam eğitim örneği sayısı
• Epoch: Bir ENTIRE veri kümesi yalnızca ONCE sinir ağı üzerinden ileri ve geri
aktarıldığında yineleme sayısı.

```python
hist = model.fit(x_train, y_train_one_hot, 
           batch_size=256, epochs=10, validation_split=0.3 )
```

Hem training
hem de
validation
verileri
için modelin
doğruluğu
görselleştirildi.

```python
plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper left')
plt.show()
```

Hem training
hem de
validation
verileri için
loss modeli
görselleştirildi.

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper right')
plt.show()
```

MODELİN TEST EDİLMESİ
Sınıflandırmak istenilen veriler
bir görüntü dosyasından
my_image değişkenine yüklenir.
imshow ile yüklenen resim
gösterildi.

```python
my_image = plt.imread("Desktop/automobile.jpg") 
```

```python
img = plt.imshow(my_image)
```

Görüntüyü depth=3 olan 32x32 piksel bir görüntüye
yeniden boyutlandırıldı (resize) ve görüntü gösterildi.

```python
from skimage.transform import resize
my_image_resized = resize(my_image, (32,32,3)) 
img = plt.imshow(my_image_resized) 
```

Sınıfların etiketini, eşdeğer sayıya karşılık gelen etiketlerin
dizinindeki bir diziye ekleyin.
Örneğin, ‘airplane’ index=0 konumunda olacaktır, çünkü sayı
olarak karşılık gelen etiket 0'dır.

```python
import numpy as np
probabilities = model.predict(np.array( [my_image_resized,] ))
```

```python
probabilities
```

En olası ilk 5 sınıfı ve karşılık gelen olasılığı yazdırır.

Bu model, verilen görüntüyü %94,13 olasılıkla bir otomobil olarak doğru
bir şekilde sınıflandırdı.

```python
number_to_class = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
index = np.argsort(probabilities[0,:])
print("Most likely class:", number_to_class[index[9]], "-- Probability:", probabilities[0,index[9]])
print("Second most likely class:", number_to_class[index[8]], "-- Probability:", probabilities[0,index[8]])
print("Third most likely class:", number_to_class[index[7]], "-- Probability:", probabilities[0,index[7]])
print("Fourth most likely class:", number_to_class[index[6]], "-- Probability:", probabilities[0,index[6]])
print("Fifth most likely class:", number_to_class[index[5]], "-- Probability:", probabilities[0,index[5]])
```
